```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## Set seaborn default theme

```python
sns.set()
```

```python
sns.set_theme()
```

```python
sns.set_style("dark")
```

## Import .CSV File data

```python
df = pd.read_csv("data.csv")

df.head()
```

## Use lmplot function to make a scatter plot

```python
sns.lmplot(x='Height',y='Weight', data=df)
```

## Create a diffrent style visualization

```python
sns.lmplot(x='Height', y='Weight', data=df,hue='Weight')
```

```python
sns.catplot(data=df, kind="bar", x="Weight", y="Height")
```

```python
sns.catplot(data=df, kind="violin", x="Weight", y="Height")
```

```python
sns.boxplot(data=df)
```
