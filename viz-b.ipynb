```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
sns.set_theme()
```

## Import and show Data tabel

```python
df = pd.read_csv("audi.csv")

df.head()
```

```python
sns.relplot(x='year', y='price', hue="year", data=df);
```

## Add price mean groub by year

```python
price_mean = df.groupby('year', as_index=False).price.mean()

price_mean.head()
```

## Create a diffrent style visualization

```python
visual = sns.barplot(x="year", y="price", data=price_mean, ci=None)
```

```python
sns.lmplot(x="engineSize", y="mpg", data=df);
```
